```python
move=pd.read_excel('1.2.3py.xlsx', sheet_name="Лист3")
move["I"]=4.04/10*2.429*(move["T,c"])**2
move["h"]=move["N,дел"]*0.5
move
```

```python
print(move.T.style.to_latex(clines="all;data"))
```

```python
import matplotlib.pyplot as plt
k,a=np.polyfit(move["h"]**2, move["I"], 1)
print(k,a)
```

```python
plt.figure(figsize=(8,4), dpi=200) # размер графика
plt.ylabel("$I \\cdot 10^3, кг\\cdot c^2 $") # подписи к осям
plt.xlabel("$h^2, см^2$")
plt.grid(True, linestyle="--") # сетка
#plt.axis([0,0.25,0,1.3]) # масштабы осей
plt.plot(move["h"]**2, k * move["h"]**2 + a, "-g",linewidth=1, label="$ I = %.2f h^2 +  %.2f$" % (k, a)) # аппроксимация
plt.scatter(move["h"]**2, move["I"], color="black",s=8, marker='o')
plt.legend() # легенда
```

```python
def sigma_mnk (x,y):
    mu = np.mean(y) # средее
    mv = np.mean(x)
    mv2 = np.mean(x**2) # средний квадрат
    mu2 = np.mean(y**2)
    muv = np.mean (y * x) # среднее от произведения
    k = (muv - mu * mv) / (mv2 - mv**2)
    b = mu - k * mv
    sigma_k = np.sqrt(1/(len(x)-2) * ( (mu2 - mu**2)/(mv2 - mv**2) - k**2 ) )
    sigma_b = sigma_k * np.sqrt(mv2)
    return (sigma_k,sigma_b)
```

```python
sigma_k,sigma_b=sigma_mnk(move["h"]**2, move["I"])
print(sigma_k,sigma_b)
```
