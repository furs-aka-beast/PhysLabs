```python
from sympy import *
from sympy.plotting import plot 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sympy.plotting.plot import MatplotlibBackend
```

```python
x, y=symbols('x y')
fi_max=pi/4
eq=2/pi/sqrt(1-sin(y/2)**2*sin(x)**2)
integrated=integrate(eq, (x, pi/2))-1
```

```python
p1=plot(integrated, (y,0,pi/2), label="    ", show=0)
p2=plot(x**2/16, (x,0,pi/2), label="    ", show=0)
p3=plot(integrated+delta**2/8*pi**2, (y,0,pi/2), label="  c  ", show=0)
p1.append(p2[0])
p1.append(p3[0])
p1.legend=True
p1.size=[2.97*5,2.1*5]
p1.xlabel=r"$ \varphi_m $"
p1.ylabel="$ \Delta T /T_0 $"
be = MatplotlibBackend(p1)
be.process_series()

plt = be.plt
plt.plot([0,1],[0,0.1],'o',markersize=1,label=" ")


# Update the legend to include markers
plt.legend()
plt.savefig('1',dpi=500)
plt.show()
```
