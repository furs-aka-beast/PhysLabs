```python
import pandas as pd
import numpy as np
data=pd.read_excel('2.2.6.xlsx', sheet_name="Лист1")
```

```python
data["v"]=data.l/100/data.tau
data["dm"]=(data.d1+data.d2)/2
data
```

```python
data["eta"]=2/9*9.8*data.dm**2/4*(data.rho_m-data.rho_l)/data.v
data["tau1"]=data.dm**2/2/9*data.rho_m/data.eta
data["S"]=data.v*data.tau1*1e5
data["Re"]=data.rho_l*1000*data.v*0.03/data.eta*1e3
data["delta_eta"]=data.eta*np.sqrt((0.02/data.dm)**2+(6e-4/data.v)**2)
data
```

```python
table=data[["d1", "d2", "dm", "tau","t", "l", "material", "eta","delta_eta", "tau1","S", "Re"]]
table
```

```python
print(table.style.to_latex(clines="all;index"))
```

```python
data["lneta"]=np.log(data.eta)
data["1/T"]=1/(data.t+273)
data["deltaln"]=data.delta_eta/data.eta
data
```

```python
k,a=np.polyfit(data["1/T"], data["lneta"], 1)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6), dpi=100) # размер графика
plt.ylabel("$ln(\eta)$") # подписи к осям
plt.xlabel("$1/T$, $1/K$")
plt.grid(True, linestyle="--") # сетка
#plt.axis([0,0.25,0,1.3]) # масштабы осей
x = np.array([0., 1]) # две точки аппроксимирующей прямой
plt.plot(data["1/T"], k * data["1/T"] + a, "-r",linewidth=1, label="Линейная аппроксимация $u = %.2f v %.2f$" % (k, a)) # аппроксимация
plt.errorbar(data[data["material"]=="s"]["1/T"], data[data["material"]=="s"]["lneta"], data[data["material"]=="s"]["deltaln"],  fmt="ob", label="Стальные шары", ms=2) # точки с погрешностями
plt.errorbar(data[data["material"]=="g"]["1/T"], data[data["material"]=="g"]["lneta"], data[data["material"]=="g"]["deltaln"],  fmt="og", label="Стекляные шары", ms=2)
plt.legend() # легенда
```
